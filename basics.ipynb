```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x_train = np.array([1.0, 2.0, 3.0, 4.0, 6.0, 8.0])

y_train = np.array([200.0, 500.0, 700.0, 900.0, 1300, 1600.0])
print(f"x_train = {x_train}")
print(f"y_train = {y_train}")
```

```python
print(f"x_train.shape: {x_train.shape}")
m = x_train.shape[0]
# we can also use len() function  instead of shape
print(f"y_train.shape: {y_train.shape}")
```

```python
i=1
x_i = x_train[i]
y_i = y_train[i]
print(f"(x^({i}), y^({i})) = ({x_i}, {y_i})")
```

```python
plt.scatter(x_train, y_train, marker="x", c='r')
plt.title("housing prices")
plt.ylabel("price (in 1000s of dollars)")
plt.xlabel("size in 1000 sqft")
plt.show()
```

```python
w=100
b=100
```

```python
def compute_model_output(x, w ,b):
  m=x.shape[0]
  f_wb=np.zeros(m)
  for i in range(m):
    f_wb[i]=w*x[i] +  b
    print(f"f_wb[{i}]: {f_wb[i]}")
  return f_wb
```

```python
real_f_wb = compute_model_output(x_train, w ,b)
plt.plot(x_train, real_f_wb, c="b", label="the prediction")

plt.scatter(x_train, y_train, marker='x', c='r',label='Actual Values')

plt.title("housing prices")

plt.ylabel('Price (in 1000s of dollars)')
# Set the x-axis label
plt.xlabel('Size (1000 sqft)')
plt.legend()
plt.show()
```
